### 8. Find the average rating for movies released in the 20th century and the average rating for movies released in the 21st century for the movies in *imdb*.

```python
#avg rating for movies in the 20th century
round(df[df['Decade'].between(1901, 2000)]['Rating'].mean(), 2)
```

```python
#avg rating for movies in the 21st century
round(df[df['Decade'] > 2000]['Rating'].mean(), 2)
```

### 9. Use number of rows and arithmetic to find the proportion of movies in the dataset that were released in the 20th century, and the proportion from the 21st century.

```python
#proportions of movies released in the 20th century
total_for_20th = len(df[df['Decade'].between(1901, 2000)])
prop_for_20th = total_for_20th/len(df)
print(round(prop_for_20th, 2))
```

```python
#proportion of movies released in the 21st century
total_for_21st = len(df[df['Decade'] > 2000])
prop_for_21st = total_for_21st/len(df)
print(round(prop_for_21st, 2))
```
